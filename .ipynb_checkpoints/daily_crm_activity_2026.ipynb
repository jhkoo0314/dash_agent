```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 1. 마스터 로직 및 배정 데이터 불러오기
try:
    # 활동 가중치 로직 로드
    df_logic = pd.read_excel(r'c:\agent_b\data\logic\SFE_Master_Logic_v1.0.xlsx', sheet_name='Activity_Weights')
    activities = df_logic[['활동명', '가중치']].to_dict('records')
    
    # 담당자-병원 배정 데이터 로드
    df_assigned = pd.read_excel("hospital_assignment_data.xlsx")
    
    # 품목 데이터 정의 (이전 단계 품목명 반영)
    blood_products = [f"혈액제제_{p}" for p in ['알부민', '아이비글로불린', '안티트롬빈', '히스토불린', '알부민-20%']]
    rx_products = ['로타칸정', '다비듀오', '이스포나인', '당뇨트리플', '인슐정', '다소미정', '아소캡', '심플루', '백신', '인수액']
    all_products = blood_products + rx_products
    
    print("✅ 모든 마스터 데이터를 성공적으로 불러왔습니다.")
except Exception as e:
    print(f"❌ 데이터 로드 중 오류 발생: {e}")

# 2. 2026년 영업일(평일) 리스트 생성
start_date = datetime(2026, 1, 1)
end_date = datetime(2026, 12, 31)
all_dates = pd.date_range(start_date, end_date)
# 주말(토, 일) 제외한 영업일만 필터링
working_days = [d for d in all_dates if d.weekday() < 5]

# 3. 가상 교수(고객) 리스트 생성 함수
doctors = ['김교수', '이교수', '박교수', '최교수', '정교수', '강교수', '조교수']

# 4. 일별 CRM 활동내역 생성
crm_logs = []

print(f"CRM 활동 데이터 생성 중 (영업일 {len(working_days)}일)...")

for date in working_days:
    date_str = date.strftime('%Y-%m-%d')
    
    # 45명의 모든 담당자에 대해 활동 생성
    for _, rep in df_assigned[['담당자ID', '담당자명', '지점ID', '지점명']].drop_duplicates().iterrows():
        
        # 해당 담당자가 관리하는 병원들 리스트
        rep_hospitals = df_assigned[df_assigned['담당자ID'] == rep['담당자ID']].to_dict('records')
        
        # 하루 활동 수 설정 (8~12콜)
        daily_call_count = random.randint(8, 12)
        
        for _ in range(daily_call_count):
            target_h = random.choice(rep_hospitals)
            target_act = random.choice(activities)
            
            crm_logs.append({
                '활동일자': date_str,
                '지점ID': rep['지점ID'],
                '지점명': rep['지점명'],
                '담당자ID': rep['담당자ID'],
                '담당자명': rep['담당자명'],
                '병원ID': target_h['병원ID'],
                '요양기관명': target_h['요양기관명'],
                '고객명': random.choice(doctors),
                '품목명': random.choice(all_products),
                '활동명': target_act['활동명'],
                '콜수': 1,
                '가중치': target_act['가중치'],
                '환산콜(Weighted)': 1 * target_act['가중치']
            })

# 5. 결과 데이터프레임 생성 및 저장
df_crm_final = pd.DataFrame(crm_logs)

# 6. 통계 요약 및 저장
print(f"\n✅ CRM 활동 데이터 생성 완료 (총 {len(df_crm_final):,}건)")
print(f"- 데이터 규모: 약 {len(df_crm_final)//10000}만 건")
print(f"- 평균 일일 활동량: {df_crm_final.groupby(['활동일자', '담당자명'])['콜수'].sum().mean():.1f} 콜")

save_path = "daily_crm_activity_2026.xlsx"
# 데이터가 너무 크면 엑셀 저장 시 시간이 걸릴 수 있어 샘플(10만건) 혹은 CSV 저장을 추천할 수 있습니다.
df_crm_final.to_excel(save_path, index=False)
print(f"🚀 파일 저장 완료: '{save_path}'")

# 결과 샘플 출력
df_crm_final.head(10)
```
