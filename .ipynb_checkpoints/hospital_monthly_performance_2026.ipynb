```python
import pandas as pd
import numpy as np
import random
import os

# 1. 파일 존재 여부 확인 및 불러오기
target_file = "hospital_monthly_targets_2026.xlsx"

if not os.path.exists(target_file):
    print(f"❌ 에러: '{target_file}' 파일이 현재 폴더에 없습니다.")
    print("👉 해결방법: 이전 단계의 '월별 목표 생성 코드'가 있는 셀을 먼저 실행해서 엑셀 파일을 만들어주세요.")
else:
    # 2. 데이터 불러오기
    df_performance = pd.read_excel(target_file)
    print(f"✅ '{target_file}' 데이터를 성공적으로 불러왔습니다. (총 {len(df_performance):,}행)")

    # 3. 실적 데이터 생성 로직 (달성률 반영)
    # 평균 95%, 표준편차 15%의 정규분포로 달성률 생성
    np.random.seed(42)
    attainment_rates = np.random.normal(loc=0.95, scale=0.15, size=len(df_performance))
    
    # 달성률 범위를 40% ~ 150% 사이로 제한
    attainment_rates = np.clip(attainment_rates, 0.4, 1.5)

    # 4. 실적 금액 계산 (정수형 변환)
    df_performance['실적금액'] = (df_performance['목표금액'] * attainment_rates).astype(int)

    # 5. 분석용 컬럼 추가
    df_performance['달성률'] = round((df_performance['실적금액'] / df_performance['목표금액']) * 100, 1)
    df_performance['목표대비차이'] = df_performance['실적금액'] - df_performance['목표금액']

    # 6. 결과 요약 출력
    print("\n[전사 실적 요약]")
    summary = df_performance.agg({'목표금액': 'sum', '실적금액': 'sum'})
    total_rate = round((summary['실적금액'] / summary['목표금액']) * 100, 1)
    print(f"전체 목표: {summary['목표금액']:,.0f}원")
    print(f"전체 실적: {summary['실적금액']:,.0f}원")
    print(f"평균 달성률: {total_rate}%")

    # 7. 파일 저장
    output_file = "hospital_monthly_performance_2026.xlsx"
    df_performance.to_excel(output_file, index=False)
    print(f"\n🚀 실적 데이터 저장이 완료되었습니다: '{output_file}'")

    # 샘플 표시
    display(df_performance.head(10))
```

```python
import pandas as pd
import numpy as np
import random
import os

# 1. ID와 품목명이 포함된 최신 목표 데이터 불러오기
target_file = "hospital_monthly_targets.xlsx"

if not os.path.exists(target_file):
    print(f"❌ 에러: '{target_file}' 파일이 없습니다. 목표 생성(v3) 코드를 먼저 실행해주세요.")
else:
    df_perf = pd.read_excel(target_file)
    print(f"✅ 목표 데이터를 불러왔습니다. (총 {len(df_perf):,}행)")

    # 2. 실적 데이터 생성 (달성률 40% ~ 150%, 평균 95% 정규분포)
    np.random.seed(42) # 분석 일관성을 위해 시드 고정
    attainment_rates = np.random.normal(loc=0.95, scale=0.15, size=len(df_perf))
    attainment_rates = np.clip(attainment_rates, 0.40, 1.50)

    # 3. 실적 계산 및 관련 컬럼 추가
    df_perf['실적금액'] = (df_perf['목표금액'] * attainment_rates).astype(int)
    df_perf['달성률(%)'] = round((df_perf['실적금액'] / df_perf['목표금액']) * 100, 1)
    df_perf['목표대비차이'] = df_perf['실적금액'] - df_perf['목표금액']

    # 4. 분석용 결과 정리 (컬럼 순서 정비)
    # 모든 ID와 이름이 나란히 배치되도록 구성합니다.
    final_cols = [
        '목표월', 
        '지점ID', '지점명', 
        '담당자ID', '담당자명', 
        '병원ID', '요양기관명', 
        '품목ID', '품목명', 
        '목표금액', '실적금액', '달성률(%)', '목표대비차이'
    ]
    df_result = df_perf[final_cols]

    # 5. 요약 정보 출력
    print("\n[ID 매칭 실적 요약]")
    total_summary = df_result.agg({'목표금액': 'sum', '실적금액': 'sum'})
    print(f"- 전체 목표액: {total_summary['목표금액']:,.0f}원")
    print(f"- 전체 실적액: {total_summary['실적금액']:,.0f}원")
    print(f"- 전사 달성률: {round(total_summary['실적금액']/total_summary['목표금액']*100, 1)}%")

    # 6. 파일 저장
    output_file = "hospital_performance_final_v3.xlsx"
    df_result.to_excel(output_file, index=False)
    print(f"\n🚀 ID 기반 최종 실적 데이터 저장 완료: '{output_file}'")

    # 샘플 보기
    display(df_result.head(10))
```
