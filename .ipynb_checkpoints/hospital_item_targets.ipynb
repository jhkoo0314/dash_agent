```python
import pandas as pd
import numpy as np
import random

# 1. 이전 단계에서 생성한 배정 데이터 불러오기
try:
    df_assigned = pd.read_excel("hospital_assignment_data.xlsx")
except FileNotFoundError:
    print("Error: 'hospital_assignment_data.xlsx' 파일이 없습니다. 이전 단계 코드를 먼저 실행해주세요.")

# 2. 가상 품목(Product) 정의 (이전과 동일)
products = []
# 혈액제제 (5개) - 가중치 합 50%
blood_products = [f"{name}" for name in ['알부민', '아이비글로불린', '안티트롬빈', '히스토불린', '알부민-20%']]
for p in blood_products:
    products.append({'품목명': p, '카테고리': '혈액제제', '세부카테고리': '혈액제제', '가중치_그룹': 0.5 / 5})

# Rx처방품목 (10개) - 가중치 합 50%
rx_cardio = [f"{i}" for i in ['로타칸정', '다비듀오', '이스포나인']]
rx_endo = [f"{i}" for i in ['당뇨트리플', '인슐정', '다소미정', '아소캡']]
rx_etc = [f"{i}" for i in ['심플루', '백신', '인수액']]

for p in rx_cardio:
    products.append({'품목명': p, '카테고리': 'Rx처방품목', '세부카테고리': '심혈관', '가중치_그룹': 0.5 / 10})
for p in rx_endo:
    products.append({'품목명': p, '카테고리': 'Rx처방품목', '세부카테고리': '내분비', '가중치_그룹': 0.5 / 10})
for p in rx_etc:
    products.append({'품목명': p, '카테고리': 'Rx처방품목', '세부카테고리': '기타', '가중치_그룹': 0.5 / 10})

df_products = pd.DataFrame(products)

# 3. 월별(26.01 ~ 26.12) 목표 데이터 생성
months = [f"2026-{str(m).zfill(2)}" for m in range(1, 13)]
reps = df_assigned['담당자명'].unique()

target_rows = []

print("데이터 생성 중... (잠시만 기다려주세요)")

for month in months:
    for rep in reps:
        # 해당 월, 해당 담당자의 월간 총 목표액 (1억 ~ 3억)
        monthly_rep_target = random.randint(100, 300) * 1000000
        
        rep_hospitals = df_assigned[df_assigned['담당자명'] == rep].copy()
        
        # 병원별 가중치 (상급종합 2.0, 종합병원 1.0)
        rep_hospitals['병원가중치'] = rep_hospitals['종별코드명'].apply(lambda x: 2.0 if x == '상급종합' else 1.0)
        total_h_weight = rep_hospitals['병원가중치'].sum()
        
        for _, h_row in rep_hospitals.iterrows():
            # 병원별 월간 할당액
            hospital_monthly_share = (h_row['병원가중치'] / total_h_weight) * monthly_rep_target
            
            for _, p_row in df_products.iterrows():
                # 품목별 월간 목표액 (혈액 50%, Rx 50% 비중 유지 + 랜덤 편차)
                monthly_item_target = hospital_monthly_share * p_row['가중치_그룹'] * random.uniform(0.9, 1.1)
                
                # --- [수정 포인트] ---
                # int()를 사용하여 소수점을 완전히 제거하고 '원' 단위 정수로 변환합니다.
                final_target_amount = int(monthly_item_target)
                
                target_rows.append({
                    '목표월': month,
                    '담당자명': rep,
                    '지점명': h_row['지점명'],
                    '요양기관명': h_row['요양기관명'],
                    '병원종류': h_row['종별코드명'],
                    '품목카테고리': p_row['카테고리'],
                    '세부카테고리': p_row['세부카테고리'],
                    '품목명': p_row['품목명'],
                    '목표금액': round(monthly_item_target, -3)
                })

df_monthly_target = pd.DataFrame(target_rows)

# 4. 데이터 검증 및 결과 확인
print(f"총 생성된 데이터 행 수: {len(df_monthly_target):,}개")
print(f"담당자별 월평균 목표액: {df_monthly_target.groupby(['목표월', '담당자명'])['목표금액'].sum().mean():,.0f}원")

# 월별 목표 총합 확인 (계절성 없이 랜덤하게 생성됨)
print("\n[월별 목표 총계 (단위: 억)]")
month_summary = df_monthly_target.groupby('목표월')['목표금액'].sum() / 100000000
print(month_summary)

# 결과 저장 (데이터 양이 많으므로 엑셀 저장 시 시간이 다소 소요될 수 있습니다)
save_path = "hospital_monthly_targets_2026.xlsx"
df_monthly_target.to_excel(save_path, index=False)
print(f"\n'{save_path}' 파일로 저장이 완료되었습니다.")

# 샘플 출력 (2026년 1월 데이터 위주)
df_monthly_target.head(15)
```

```python
import pandas as pd
import numpy as np
import random
import os

# 1. 이전 단계에서 생성한 ID 기반 배정 데이터 불러오기
assignment_file = "hospital_assignment_data.xlsx"
if not os.path.exists(assignment_file):
    print(f"❌ 에러: '{assignment_file}' 파일이 없습니다. 배정 단계(v2) 코드를 먼저 실행해주세요.")
else:
    df_assigned = pd.read_excel(assignment_file)

    # 2. 품목(Product) 정의 및 요청하신 이름 반영
    products = []
    
    # [혈액제제] - 가중 합 50% (P001 ~ P005)
    blood_names = ['알부민', '아이비글로불린', '안티트롬빈', '히스토불린', '알부민-20%']
    for i, name in enumerate(blood_names):
        products.append({
            '품목ID': f"P{str(i+1).zfill(3)}",
            '품목명': f"혈액제제_{name}",
            '카테고리': '혈액제제',
            '가중치': 0.5 / 5
        })

    # [Rx처방품목] - 가중 합 50% (P006 ~ P015)
    # 요청하신 이름으로 카테고리별 구성
    rx_cardio = ['로타칸정', '다비듀오', '이스포나인'] # 심혈관 3
    rx_endo = ['당뇨트리플', '인슐정', '다소미정', '아소캡'] # 내분비 4
    rx_etc = ['심플루', '백신', '인수액'] # 기타 3
    
    all_rx_names = rx_cardio + rx_endo + rx_etc
    
    for i, name in enumerate(all_rx_names):
        products.append({
            '품목ID': f"P{str(i+6).zfill(3)}",
            '품목명': name,
            '카테고리': 'Rx처방품목',
            '가중치': 0.5 / 10
        })

    df_products = pd.DataFrame(products)

    # 3. 월별 목표 생성 루프 (26.01 ~ 26.12)
    months = [f"2026-{str(m).zfill(2)}" for m in range(1, 13)]
    target_rows = []

    print("변경된 품목명 반영 및 목표 생성 중...")

    for month in months:
        unique_reps = df_assigned[['지점ID', '지점명', '담당자ID', '담당자명']].drop_duplicates()
        
        for _, rep_row in unique_reps.iterrows():
            monthly_rep_target = random.randint(100, 300) * 1000000
            rep_hospitals = df_assigned[df_assigned['담당자ID'] == rep_row['담당자ID']].copy()
            rep_hospitals['병원가중치'] = rep_hospitals['종별코드명'].apply(lambda x: 2.0 if x == '상급종합' else 1.0)
            total_h_weight = rep_hospitals['병원가중치'].sum()
            
            for _, h_row in rep_hospitals.iterrows():
                hospital_share = (h_row['병원가중치'] / total_h_weight) * monthly_rep_target
                
                for _, p_row in df_products.iterrows():
                    item_target = hospital_share * p_row['가중치'] * random.uniform(0.9, 1.1)
                    
                    target_rows.append({
                        '목표월': month,
                        '지점ID': rep_row['지점ID'],
                        '지점명': rep_row['지점명'],
                        '담당자ID': rep_row['담당자ID'],
                        '담당자명': rep_row['담당자명'],
                        '병원ID': h_row['병원ID'],
                        '요양기관명': h_row['요양기관명'],
                        '병원종류': h_row['종별코드명'],
                        '품목ID': p_row['품목ID'],
                        '품목명': p_row['품목명'],
                        '품목카테고리': p_row['카테고리'],
                        '목표금액': int(item_target)
                    })

    df_target_final = pd.DataFrame(target_rows)

    # 4. 결과 정리 및 저장
    cols = ['목표월', '지점ID', '지점명', '담당자ID', '담당자명', '병원ID', '요양기관명', '품목ID', '품목명', '목표금액']
    df_result = df_target_final[cols]

    save_path = "hospital_monthly_targets_v3_named.xlsx"
    df_result.to_excel(save_path, index=False)
    
    print(f"✅ 완료! 총 {len(df_result):,}행 데이터가 생성되었습니다.")
    print(f"🚀 저장 파일명: '{save_path}'")

    # 결과 샘플 (Rx 품목 위주로 확인)
    display(df_result[df_result['품목ID'].isin(['P006', 'P009', 'P013'])].head(10))
```
